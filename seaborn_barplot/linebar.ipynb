```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset("diamonds")
df
```

## ðŸ”´You can draw a bar plot using pandas.

```python
table = pd.crosstab(df["cut"], columns="count")
table
```

```python
table.plot(kind="bar")
```

## ðŸ”´You can also use the seaborn library and catplot to draw a bar chart.

```python
sns.catplot(data=df, x="cut", kind="count")
```

## ðŸ”´Changing the aspect changes the width of the rectangles.

```python
sns.catplot(data=df, x="cut", kind="count", aspect=3)
```

## ðŸ”´You can change the color and palette.

```python
sns.catplot(data=df, x="cut", kind="count", color="m") 
```

```python
sns.catplot(data=df, x="cut", kind="count", palette="dark", hue="cut")
```

## ðŸ”´The order of the columns is also adjustable.

```python
sns.catplot(data=df, x="cut", kind="count", palette="dark", hue="cut",
           order=['Ideal', 'Fair','Premium', 'Good', 'Very Good'])
```

## ðŸ”´Sets the seaborn style to a white grid background.

```python
sns.set(style="whitegrid")
```

```python
sns.barplot(data=df, x="cut", y="price")
```

## ðŸ”´You can change the estimator type from the default mean to, for example, median.

```python
sns.barplot(data=df, x="cut", y="price",estimator=np.median)
```

## ðŸ”´You can specify the hue.

```python
sns.barplot(data=df, x="cut", y="price",estimator=np.mean, hue="color")
```

## ðŸ”´You can swap the x and y.

```python
sns.barplot(data=df, y="cut", x="price",estimator=np.mean, orient="h")
```

## ðŸ”´You can configure the errorbar according to your needs.

```python
sns.barplot(data=df, x="cut", y="price",estimator=np.mean, errorbar=None)
```

```python
sns.barplot(data=df, x="cut", y="price",estimator=np.mean, errorbar=("ci",99))
```

```python
sns.barplot(data=df, x="cut", y="price",estimator=np.mean, errorbar=("ci",99), capsize=0.3)
```

## ðŸ”´The box for hues has customization capabilities, examples of which are provided below.

```python
ax = sns.barplot(data=df, x="cut", y="price", hue="color")
ax.legend(title="COLOR")
```

```python
ax = sns.barplot(data=df, x="cut", y="price", hue="color")
ax.legend(title="COLOR", loc="upper left")
```

```python
ax = sns.barplot(data=df, x="cut", y="price", hue="color")
ax.legend(title="COLOR", loc="upper left", fontsize=20)
```

```python
ax = sns.barplot(data=df, x="cut", y="price", hue="color")
ax.legend(title="COLOR", loc="upper left", fontsize=20, title_fontsize=50)
```

```python
ax = sns.barplot(data=df, x="cut", y="price", hue="color")
ax.legend(title="COLOR", loc="upper left",ncol=5)
```

## ðŸ”´You can change the y-label and x-label.

```python
ax = sns.barplot(data=df, x="cut", y="price", hue="color")
ax.set_ylabel("PRICE", fontsize=20)
ax.set_xlabel("CUT", fontsize=20)
```

## ðŸ”´Sets the x-axis tick labels with a font size of 10 and rotates them 45 degrees.

```python
ax = sns.barplot(data=df, x="cut", y="price", hue="color")
ax.set_xticklabels(ax.get_xticklabels(), fontsize=10, rotation=45)
```

## ðŸ”´Enables grid lines for both axes with a dashed line style and 60% opacity.

```python
ax = sns.barplot(data=df, x="cut", y="price", hue="color")
ax.grid(True, axis="both", linestyle="--", alpha=0.6)
```
