```python
import seaborn as sns
import plotly.express as px
```

```python
df = sns.load_dataset("iris")
df
```

### ðŸ”´Code Explanation
This code uses the `plotly.express` library to create a scatter plot  
- `data_frame=df` specifies the input dataset  
- `x="sepal_length"` sets the column for the x-axis  
- `y="sepal_width"` sets the column for the y-axis  
Finally, `fig.show()` displays the generated plot


```python
fig = px.scatter(data_frame=df, x="sepal_length", y="sepal_width")
fig.show()
```

- `title="Scatter plot"` sets the title of the chart 

```python
fig = px.scatter(data_frame=df, x="sepal_length", y="sepal_width", title="Scatter plot", )
fig.show()
```

- `color="species"` assigns different colors to points based on the `species` column  

```python
fig = px.scatter(data_frame=df, x="sepal_length", y="sepal_width", title="Scatter plot", 
                 color="species")
fig.show()
```

- `template="plotly_dark"` applies a dark theme to the plot  

```python
fig = px.scatter(data_frame=df, x="sepal_length", y="sepal_width", title="Scatter plot", 
                 color="species", template="plotly_dark")
fig.show()
```

- `hover_name="species"` displays the species name when hovering over a point  
- `hover_data=["petal_length"]` shows additional information (petal length) on hover  

```python
fig = px.scatter(data_frame=df, x="sepal_length", y="sepal_width", title="Scatter plot", 
                 color="species", hover_name="species",
                hover_data=["petal_length"])
fig.show()
```

- `size="petal_length"` sets the size of each point based on the `petal_length` column  
- `size_max=20` limits the maximum marker size to 20  


```python
fig = px.scatter(data_frame=df, x="sepal_length", y="sepal_width", title="Scatter plot", 
                 color="species", hover_name="species",
                hover_data=["petal_length"], size="petal_length", size_max=20)
fig.show()
```

- `symbol="species"` assigns different marker symbols to points based on the `species` column  


```python
fig = px.scatter(data_frame=df, x="sepal_length", y="sepal_width", title="Scatter plot", 
                 symbol="species")
fig.show()
```

- `width=700` sets the plot width to 700 pixels  
- `height=400` sets the plot height to 400 pixels  


```python
fig = px.scatter(data_frame=df, x="sepal_length", y="sepal_width", title="Scatter plot", 
                 color="species", width=700, height=400)
fig.show()
```

- `labels` customizes the axis and legend labels:  
  - `"sepal_length":"Sepal Length (cm)"` sets a readable label for the x-axis  
  - `"sepal_width":"Sepal Width (cm)"` sets a readable label for the y-axis  
  - `"species":"Species Flowers"` renames the legend for the `species` column  


```python
fig = px.scatter(data_frame=df, x="sepal_length", y="sepal_width", title="Scatter plot", 
                 color="species",
                labels={
                   "sepal_length":"Sepal Length (cm)",
                   "sepal_width":"Sepal Width (cm)",
                    "species":"Species Flowers"}
                )
fig.show()
```

- `update_layout(title=...)` customizes the plot title:  
  - `"x":0.5` centers the title horizontally  
  - `"text":"Scatter Plottt"` sets the title text  
  - `"font":{"size":20, "color":"blue"}` sets the title font size and color  


```python
fig.update_layout(title={
    "x":0.5,
    "text":"Scatter Plottt",
    "font":{"size":20, "color":"blue"}
    } 
                 )
```

- `xaxis` customizes the x-axis:  
  - `"title":"lenght"` sets the axis title  
  - `"range":[-10,10]` defines the visible range of the axis  
  - `"zerolinecolor":"black"` colors the zero line in black  

- `yaxis` customizes the y-axis:  
  - `"tickangle":45` rotates the tick labels by 45 degrees  
  - `"range":[-4,5]` defines the visible range of the axis  
  - `"zerolinecolor":"black"` colors the zero line in black  


```python
fig.update_layout(
    xaxis={
        "title":"lenght",
        "range":[-10,10],
        "zerolinecolor":"black",
    },
    yaxis={
        "tickangle":45,
        "range":[-4,5],
        "zerolinecolor":"black"
    }
                 )
```

```python
fig = px.scatter(data_frame=df, x="sepal_length", y="sepal_width", title="Scatter plot", 
                 color="species",
                labels={
                   "sepal_length":"Sepal Length (cm)",
                   "sepal_width":"Sepal Width (cm)",
                    "species":"Species Flowers"}
                )
fig.show()
```

- `legend` customizes the plot legend:  
  - `"orientation":"h"` arranges the legend horizontally  
  - `"x":0.5` centers the legend horizontally  
  - `"y":1` positions the legend at the top of the plot  
  - `"bgcolor":"black"` sets the legend background color to black  


```python
fig.update_layout(
    legend = {"orientation":"h",
              "x":0.5,
              "y":1,
              "bgcolor":"black"
         }
)
    
```

- `margin` sets the spacing around the plot:  
  - `"l":100` left margin  
  - `"r":100` right margin  
  - `"t":100` top margin  
  - `"b":100` bottom margin  


```python
fig.update_layout(
    margin={
        "l":100,
        "r":100,
        "t":100,
        "b":100,
    }
)
```

- `paper_bgcolor="lightyellow"` sets the background color of the entire figure  
- `plot_bgcolor="white"` sets the background color of the plotting area  


```python
fig.update_layout(
    paper_bgcolor = "lightyellow",
    plot_bgcolor = "white"
)
```

- `xaxis={"showline":True, "linecolor":"black"}` displays the x-axis line and sets its color to black  


```python
fig.update_layout(
    xaxis = {"showline":True,
             "linecolor":"black",
            }
)
```

```python
fig = px.scatter(data_frame=df, x="sepal_length", y="sepal_width", title="Scatter plot", 
                 color="species")
fig.show()
```

- `xaxis={"showline":True, "linecolor":"black", "rangeslider":{"visible":True}}` customizes the x-axis:  
  - `showline=True` displays the axis line  
  - `linecolor="black"` sets the axis line color  
  - `rangeslider={"visible":True}` adds an interactive range slider below the x-axis  


```python
fig.update_layout(
    xaxis = {"showline":True,
             "linecolor":"black",
             "rangeslider":{"visible":True},
            }
)
```

- `add_shape(type="rect", ...)` adds a rectangle shape to the plot:  
  - `x0=5, y0=2` sets the bottom-left corner  
  - `x1=6, y1=4` sets the top-right corner  
  - `fillcolor="lightblue"` sets the rectangle color  
  - `opacity=0.5` makes the rectangle semi-transparent  


```python
fig.add_shape(
    type="rect",
    x0=5,y0=2,
    x1=6,y1=4,
    fillcolor="lightblue",
    opacity=0.5
)
```

- `add_annotation(...)` adds a text annotation to the plot:  
  - `text="this is special point"` sets the annotation text  
  - `showarrow=True` displays an arrow pointing to the specified coordinates  
  - `x=5, y=2.5` positions the annotation on the plot  
  - `bgcolor="lightblue"` sets the background color of the annotation  


```python
fig.add_annotation(
    text="this is special point",
    showarrow=True,
    x=5,
    y=2.5,
    bgcolor="lightblue",
)
```
