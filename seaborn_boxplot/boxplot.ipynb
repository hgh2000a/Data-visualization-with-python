```python
import seaborn as sns 
```

```python
tips = sns.load_dataset("tips")
tips
```

## ðŸ”´you can use seaborn to draw a box plot.

```python
sns.boxplot(x="day", y="tip", data=tips)
```

## ðŸ”´You can specify the hue.

```python
sns.boxplot(x="day", y="tip", data=tips, hue="sex")
```

## ðŸ”´You can specify the palette.

```python
sns.boxplot(x="day", y="tip", data=tips, palette="coolwarm", hue="sex")
```

## ðŸ”´You can change box width.

```python
sns.boxplot(x="day", y="tip", data=tips, width=1)
```

## ðŸ”´You can specify the line width.

```python
sns.boxplot(x="day", y="tip", data=tips, linewidth=5)
```

## ðŸ”´You use showfliers=False to create a cleaner plot that focuses on the main body of the data (the quartiles and median) without the visual distraction of extreme values.


```python
sns.boxplot(x="day", y="tip", data=tips, showfliers= False)
```

## ðŸ”´You can change the orientaion.

```python
sns.boxplot(x="tip", y="day", data=tips, orient="h")
```

## ðŸ”´You can specify the hue.

```python
sns.boxplot(x="day", y="tip", data=tips,notch=True)
```

```python
sns.boxplot(x="day", y="tip", data=tips, medianprops={"color":"white"})
```

```python
sns.boxplot(x="day", y="tip", data=tips, showmeans=True,meanline=True )
```
