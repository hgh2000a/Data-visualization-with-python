```python
import seaborn as sns 
```

```python
mpg = sns.load_dataset("mpg")
mpg
```

## ðŸ”´This data is not suitable for a scatter plot, so we move on to a line plot.

```python
sns.scatterplot(data=mpg, x="model_year", y="horsepower")
```

```python
sns.lineplot(data=mpg, x="model_year", y="horsepower")
```

## ðŸ”´You can specify the ci (confidence interval) as desired.

```python
sns.lineplot(data=mpg, x="model_year", y="horsepower", errorbar=("ci",20))
```

## ðŸ”´You can change the line width.

```python
sns.lineplot(data=mpg, x="model_year", y="horsepower", errorbar=("ci",20),linewidth=5)
```

```python
flights = sns.load_dataset("flights")
flights
```

## ðŸ”´You can specify the hue.

```python
sns.lineplot(data=flights, x="year", y="passengers", errorbar=("ci",0),hue="month")
```

## ðŸ”´An example for plotting a chart for three specific months.

```python
months = ["Jan", "Jun", "Dec"]
for month in months:
    month_flight = flights[flights["month"] == month]
    sns.lineplot(data=month_flight, x="year", y="passengers", errorbar=("ci",0),label=month)
```
