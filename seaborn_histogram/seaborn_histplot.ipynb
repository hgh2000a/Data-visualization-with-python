```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset("diamonds")
```

```python
df
```

## ðŸ”´You can directly use the pandas library to plot a histogram.

```python
df.hist(column="carat")
```

## ðŸ”´You can change the bins. Its default is 10.

```python
df.hist(column="carat", bins=5)
```

## ðŸ”´You can use the seaborn library to plot a histogram.

```python
sns.histplot(data=df, x="carat")
```

## ðŸ”´The bins are selected automatically, but they can also be changed manually.

```python
sns.histplot(data=df, x="carat", bins=10)
```

## ðŸ”´You can set the kde parameter to True.

```python
sns.histplot(data=df, x="carat", kde=True)
```

## ðŸ”´You can assign custom labels and choose a name for the plot.


```python
ax = sns.histplot(data=df, x="carat", kde=True)
ax.set(title="histogram of carat", xlabel="Carat", ylabel="Count")
```

## ðŸ”´You use the color parameter to change the color.

```python
ax = sns.histplot(data=df, x="carat", kde=True, color="red")
ax.set(title="histogram of carat", xlabel="Carat", ylabel="Count")
```

## ðŸ”´You can specify the hue.

```python
ax = sns.histplot(data=df, x="carat", kde=True, color="red", hue="cut")
ax.set(title="histogram of carat", xlabel="Carat", ylabel="Count")
```

## ðŸ”´You can specify different elements that change the display, for example, poly has been drawn.


```python
ax = sns.histplot(data=df, x="carat", kde=True, color="red", element="poly")
ax.set(title="histogram of carat", xlabel="Carat", ylabel="Count")
```

## ðŸ”´The x and y values are limited using xlim and ylim.

```python
ax = sns.histplot(data=df, x="carat", element="poly")
ax.set(title="histogram of carat", xlabel="Carat", ylabel="Count",xlim=(0,3),ylim=(0,4000))
```

## ðŸ”´You can specify the x and y axis divisions yourself.

```python
ax = sns.histplot(data=df, x="carat")
ax.set(title="histogram of carat", xlabel="Carat", ylabel="Count",xticks=(1,2),yticks=(2000,5000))
```

## ðŸ”´You can label each section of the x and y axes.

```python
ax = sns.histplot(data=df, x="carat")
ax.set(title="histogram of carat", xlabel="Carat", ylabel="Count",xticks=(1,2),yticks=(2000,5000),yticklabels=("bad","good"))
```
