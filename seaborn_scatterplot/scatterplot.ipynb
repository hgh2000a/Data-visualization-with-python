```python
import seaborn as sns 
```

```python
df = sns.load_dataset("mpg")
```

```python
df
```

## ðŸ”´you can use seaborn to draw a scatter plot.

```python
sns.scatterplot(data=df, x="horsepower", y="mpg")
```

## ðŸ”´You can specify the hue.

```python
sns.scatterplot(data=df, x="horsepower", y="mpg", hue="origin")
```

## ðŸ”´You can define a size for it; for example, the circle becomes larger as the weight increases.

```python
sns.scatterplot(data=df, x="horsepower", y="mpg", size="weight")
```

## ðŸ”´You can define a style for it; for example, it assigns a specific shape to each country.

```python
sns.scatterplot(data=df, x="horsepower", y="mpg", style="origin")
```
