```python
import seaborn as sns
```

```python
flights = sns.load_dataset("flights")
flights
```

## ðŸ”´Creating a Pivot Table

```python
flights_pivot = flights.pivot(index="month", columns="year", values="passengers")
flights_pivot
```

## ðŸ”´Creating a Heatmap

```python
sns.heatmap(data=flights_pivot)
```

## ðŸ”´This key parameter stands for "annotate". It instructs the function to write the data value (the actual number of passengers) inside each cell of the heatmap.

```python
sns.heatmap(data=flights_pivot, annot=True)
```

## ðŸ”´Formats the annotated numbers as integers (decimals), removing any fractional part. This is crucial for count data like passengers, ensuring clean, whole-number labels.

```python
sns.heatmap(data=flights_pivot, annot=True, fmt="d")
```

## ðŸ”´This is a keyword arguments parameter for the annotations. It passes a dictionary of styling options to the text elements. Here, it sets the text color to black.

```python
sns.heatmap(data=flights_pivot, annot=True, fmt="d", annot_kws={"color":"black"})
```

## ðŸ”´cmap="coolwarm": This parameter changes the color map (the palette used to represent data values). The "coolwarm" palette typically uses blue for low values, white for mid-range, and red for high values.

```python
sns.heatmap(data=flights_pivot, annot=True, fmt="d", cmap="coolwarm")
```

## ðŸ”´linewidths=5: This parameter sets the width of the lines that will be drawn between each cell in the heatmap, creating a distinct grids.

## ðŸ”´linecolor="pink": This parameter defines the color of those grid lines.

```python
sns.heatmap(data=flights_pivot, annot=True, fmt="d", cmap="coolwarm",linewidths=5,linecolor="pink")
```

## ðŸ”´vmin=200: This parameter sets the minimum value of the color scale. Any data value at or below 200 will be mapped to the lowest color in the palette (e.g., dark blue in "coolwarm").).

## ðŸ”´vmax=500: This parameter sets the maximum value of the color scale. Any data value at or above 500 will be mapped to the highest color in the palette (e.g., dark red in "coolwarm").

```python
sns.heatmap(data=flights_pivot, annot=True, fmt="d", cmap="coolwarm", vmin=200, vmax=500)
```

## ðŸ”´Removing the Color Bar

```python
sns.heatmap(data=flights_pivot, annot=True, fmt="d", cmap="coolwarm", cbar=False)
```

## ðŸ”´Customizing the Color Bar

```python
sns.heatmap(data=flights_pivot, annot=True, fmt="d", cmap="coolwarm", cbar_kws={"orientation":"horizontal", "shrink":8})
```

## ðŸ”´center=600: This parameter defines the data value that will be mapped to the central color (typically white) in a diverging colormap like "coolwarm". The color scale diverges from this central point.

```python
sns.heatmap(data=flights_pivot, annot=True, fmt="d", cmap="coolwarm", center=600)
```

## ðŸ”´Adjusting Heatmap Transparency

```python
sns.heatmap(data=flights_pivot, annot=True, fmt="d", cmap="coolwarm", alpha=0.2)
```

## ðŸ”´This function from the Seaborn library plots a hierarchically-clustered heatmap

```python
sns.clustermap(data=flights_pivot)
```

```python
sns.clustermap(data=flights_pivot, row_cluster=False)
```

## ðŸ”´Linkage method for clustering.	'average', 'ward', 'single'

```python
sns.clustermap(data=flights_pivot, method="ward")
```

## ðŸ”´Distance metric for clustering.	'euclidean', 'correlation'

```python
sns.clustermap(data=flights_pivot, metric="cosine")
```

## ðŸ”´Normalizes data by row (0) or column (1).

```python
sns.clustermap(data=flights_pivot, metric="cosine", z_score=1)
```

## ðŸ”´Standardizes data by row (0) or column (1).

```python
sns.clustermap(data=flights_pivot, metric="cosine", standard_scale=1)
```
